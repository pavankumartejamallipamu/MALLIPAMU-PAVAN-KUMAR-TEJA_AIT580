```python
#Q1: Find the number of entries/rows and columns in the data.
print("Number of entries/rows :",data.shape[0])
print("Number of columns :",data.shape[1])
#Q2: What is the average Monthly Income?
print("average Monthly Income : ",data['MonthlyIncome'].mean())
#Q3: What is the highest amount of HourlyRate ?
print("highest amount of Hourly Rate ",max(data['HourlyRate']))
#Q4: What is the Department, JobRole, MaritalStatus and OverTime of EmployeeNumber 10?
search_employee_id = 10
res = data[data.EmployeeNumber == search_employee_id]
if not res.empty:
    print("Department: ", format(res['Department'].values[0]))
    print("JobRole: ", res['JobRole'].values[0])
    print("MaritalStatus: ", res['MaritalStatus'].values[0])
    print("OverTime: ", res['OverTime'].values[0])
else:
    print("employee number not found: ", search_employee_id)

#Q5: What is the Employee ID of highest MonthlyIncome paid employee?
highest_Monthlyincome=data.nlargest(1,"MonthlyIncome")
print("Employee Id of highest Monthly income:",highest_Monthlyincome.EmployeeNumber.to_string(index=False))

#Q6: What is the average(mean) DailyRate group by Age for all Employees whose age is greater than 58. (hint: use groupby function)
data1=data[data['Age']>58].groupby(['Age']).mean()['DailyRate']
print(data1)



#Q7: How many unique EducationField are there?
unique_education_fields = {}

for unique_field in data['EducationField']:

 if unique_field in unique_education_fields:

  unique_education_fields[unique_field]+=1

 else:

  unique_education_fields[unique_field] = 1
print(unique_education_fields)
print("unique eduaction fields :",len(unique_education_fields))

#Q8: What are the top 5 most common JobRole?
mostcommon_job_roles = {}

for jobrole in data['JobRole']:

 if jobrole in mostcommon_job_roles:

  mostcommon_job_roles[jobrole]+=1

 else:

  mostcommon_job_roles[jobrole] = 1

print("Most Comman Job Roles: ")

mostcommon_job_roles = sorted(mostcommon_job_roles,reverse=True)

count = 0

for i in mostcommon_job_roles:

 count+=1

 print(i)

 if count == 5:

  break

#Q9: How many JobRoles represented by less than 100 employees?
print(data['JobRole'].value_counts())



#Q10: What is the correlation between Education and JobSatisfaction
print("correlation between Education and JobSatisfaction:",data['Education'].corr(data['JobSatisfaction']))



```
